```python

# AN AI FOLDER PROGRAM  using BINARY TREE.

from types import NoneType
from collections import deque

# this node represents a single node in the binary tree
class Node:
  def __init__(self, data):
    self.data= data
    self.left_child = None
    self.right_child = None

# implements the binary tree structure and its operations
class BinaryTree:
  def __init__(self):
    self.root = None


# function for insertion , for inserting new node into the tree
  def insertion(self, node, file_name,path=None):
    if node is None:
      return Node(file_name)

    if path == "left":
      node.left_child = self.insertion(node.left_child, file_name, path)
    else:
      node.right_child = self.insertion(node.right_child, file_name, path)
    return node

# function for search which performs a breadth first search to find a node

  def search(self, file_name):
        if self.root is None:
            return None

       # BFS used because of its minimum number of operations in determining the node , given the file name
        queue = deque([self.root])
        while queue:
            node = queue.popleft()
            if node.data == file_name:
                return node
            if node.left_child:
                queue.append(node.left_child)
            if node.right_child:
                queue.append(node.right_child)
        return None



  #this function deletes a node given the file name from the tree
#1. we start we deleting a node or we could a say a leaf node , since it has no a child ,hence its deletion is not complicated
  def delete(self, node , file_name):
    if node is None:
      return node

    if node.data == file_name:
      if node.left_child is None :
        return node.right_child
      elif node.right_child is None:
        return node.left_child
      temp = self.find_min(node.right_child)
      node.data = temp.data
      node.right_child = self.delete(node.right_child, temp.data)


    else:
      node.left_child = self.delete(node.left_child, file_name)
      node.right_child = self.delete(node.right_child, file_name)
    return node

#2 deletion of a node with a single child , quite complex .

  def delete_one_child(self, node, file_name):
    if node is None:
      return node

    # Find the node to delete
    if node.data == file_name:

        # Node with only one child or no child
        if node.left_child is None:
            temp = node.right_child
            root = None
            return temp
        elif node.right_child is None:
            temp = node.left_child
            root = None
            return temp
    else:
        # Search in left or right subtree
        node.left_child = self.delete_one_child(node.left_child, file_name)
        node.right_child = self.delete_one_child(node.right_child, file_name)

    return node



#this is complex process in deleting a node with two children

  def delete_two_child(self, node, file_name):
     if node is None:
        return node

    # Find the node to delete
     if file_name == node.data:
        # Node with two children
        if node.left_child and node.right_child:
            # Find inorder successor (smallest in the right subtree)
            temp = self.find_min_value_node(node.right_child)
            # Copy the inorder successor's data to this node
            node.data = temp.data
            # Delete the inorder successor
            node.right_child = self.delete_two_child(node.right_child, temp.data)
        else:  # Node with one or no child
            if node.left_child is None:
                temp = node.right_child
                root = None
                return temp
            elif node.right_child is None:
                temp = node.left_child
                node = None
                return temp
     else:
        # Search in left or right subtree
        node.left_child = self.delete_two_child(node.left_child, file_name)
        node.right_child = self.delete_two_child(node.right_child, file_name)

     return node



    #extras

  def inorder_traversal(self, node):
        if node:
            self.inorder_traversal(node.left_child)
            print(node.data, end=" ")
            self.inorder_traversal(node.right_child)


  def level_order_traversal(self, node):
        if node is None:
            return

        queue = deque([node])
        while queue:
            curr = queue.popleft()
            print(curr.data, end=" ")

            if curr.left_child:
                queue.append(curr.left_child)
            if curr.right_child:
                queue.append(curr.right_child)

  def preorder_traversal(self, node):
        if node:
            print(node.data, end=" ")
            self.preorder_traversal(node.left_child)
            self.preorder_traversal(node.right_child)


  def postorder_traversal(self, node):
        if node:
            self.postorder_traversal(node.left_child)
            self.postorder_traversal(node.right_child)
            print(node.data, end=" ")




  def print_tree(self, node=None, level=0, prefix = "Root: "):
    if node is None:
       node = self.root

    if node:

       print(" " * (level * 4) + prefix + node.data)
       if node.left_child:
                self.print_tree(node.left_child, level + 1, "L-- ")
       if node.right_child:
                self.print_tree(node.right_child, level + 1, "R-- ")

tree =BinaryTree()
tree.root = tree.insertion(tree.root, "Artificial Intelligence")
tree.insertion(tree.root,"Datasets","left")
tree.insertion(tree.root, "Models", "right")
tree.insertion(tree.root.left_child,"Diseases","left")
tree.insertion(tree.root.left_child, "Others","right")
tree.insertion(tree.root.right_child, "Regression", "left")
tree.insertion(tree.root.right_child, "Classification", "right")
tree.insertion(tree.root.left_child.left_child, "Diabetes", "left")
tree.insertion(tree.root.left_child.left_child, "Cancer", "right")
tree.insertion(tree.root.left_child.right_child, "Finance", "left")
tree.insertion(tree.root.left_child.right_child, "Education", "right")
tree.insertion(tree.root.right_child.left_child, "Linear Regression", "left")
tree.insertion(tree.root.right_child.left_child, "Logistic Regrression", "right")
tree.insertion(tree.root.right_child.right_child, "Decision Tree", "left")
tree.insertion(tree.root.right_child.right_child, "Random Forest", "right")


tree.print_tree()

search_results = [
    tree.search("Datasets"),
    tree.search("Machine Learning"),
    tree.search("Diseases"),
    tree.search("Deep Learning"),
    tree.search("Diabetes")
]

for result in search_results:
    if result is not None:
        print("\n Found : ", result.data)
    else:
        print("\n NOT FOUND")



print("\n Deleting Random Forest'...")
tree.root = tree.delete(tree.root, "Random Forest")
print("\n After Deletion :")
tree.print_tree()

print("\n Deleting Classification'...")
tree.root = tree.delete_one_child(tree.root, "Classification")
print("\n After Deletion :")
tree.print_tree()


print("\n Deleting Others'...")
tree.root = tree.delete_two_child(tree.root, "Others")
print("\n After Deletion :")
tree.print_tree()



print("\n")


print("this is inorder treversal")
tree.inorder_traversal(tree.root)
print(" \n")

print("this is preorder treversal")
tree.preorder_traversal(tree.root)
print("\n")

print("this is postorder treversal")
tree.postorder_traversal(tree.root)
print("\n")

print("this is level order treversal")
tree.level_order_traversal(tree.root)
```
